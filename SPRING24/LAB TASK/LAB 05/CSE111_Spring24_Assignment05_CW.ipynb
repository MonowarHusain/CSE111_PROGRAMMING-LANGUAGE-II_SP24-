```python
#TASK02
class Farmer:
    def __init__(self, *args):
        if len(args) == 0:
            print("Welcome to your farm!")
        else:
            if isinstance(args[0], str):
                print(f'Welcome to your farm, {args[0]}!')
            elif isinstance(args[0], int):
                print(f'Welcome to your farm. Your farm ID is {args[0]}!')
        self.crop_list = []
        self.fish_list = []

    def addCrops(self, *args):
        if len(args) == 0:
            print("No crops added.")
        else:
            print(f'{len(args)} crop(s) added.')
            for crop in args:
                self.crop_list.append(crop)

    def addFishes(self, *args):
        if len(args) == 0:
            print("No fish added.")
        else:
            print(f'{len(args)} fish(es) added.')
            for fish in args:
                self.fish_list.append(fish)

    def showGoods(self):
        if len(self.crop_list) == 0:
            print("You don't have any crops.")
        else:
            print(f'You have {len(self.crop_list)} crop(s):')
            print(", ".join(self.crop_list))
        if len(self.fish_list) == 0:
            print("You don't have any fish.")
        else:
            print(f'You have {len(self.fish_list)} fish(es):')
            print(", ".join(self.fish_list))

f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#TASK05
class Sphere:
    def __init__(self, sphere_id, radius=1, color="White"):
        self.sphere_id = sphere_id
        self.radius = radius
        self.color = color
        self.volume = self.calculate_volume()

    def calculate_volume(self):
        return 4/3 * 3.1416 * (self.radius ** 3)

    def printDetails(self):
        print(f"Sphere ID: {self.sphere_id}")
        print(f"Color: {self.color}")
        print(f"Volume: {self.volume}")

    def merge_sphere(self, *spheres):
        print("Spheres are being merged")
        total_volume = self.volume
        mixed_color = False

        for sphere in spheres:
            total_volume += sphere.volume
            if sphere.color != self.color:
                mixed_color = True

        if mixed_color:
            self.color = "Mixed Color"

        self.radius = ((3 * total_volume) / (4 * 3.1416)) ** (1/3)
        self.volume = total_volume

# Driver Code
sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1, sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```
