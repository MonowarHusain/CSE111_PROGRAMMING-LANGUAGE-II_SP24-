```python
#TASK_01

class Customer:
    def __init__(self):
        self.tickets = []
        self.max_tickets = 3
        print("Welcome to ABC Memorial Park")

    def buyTicket(self, name, age):

        if len(self.tickets) < self.max_tickets:
            if age > 10:
                self.tickets.append(100)
            else:
                self.tickets.append(50)
            print(f"Successfully purchased a ticket for {name}!")
        else:
            print("You can't buy more than 3 tickets")

    def showDetails(self):
        print("Amount of tickets:", len(self.tickets))
        print("Total price:", sum(self.tickets), "Taka")


print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
#TASK_02
class Panda:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age

    def sleep(self, hours):
        if 3 <= hours <= 5:
            return "{} sleeps {} hours daily and should have Mixed Veggies".format(self.name, hours)
        elif 6 <= hours <= 8:
            return "{} sleeps {} hours daily and should have Eggplant & Tofu".format(self.name, hours)
        elif 9 <= hours <= 11:
            return "{} sleeps {} hours daily and should have Broccoli Chicken".format(self.name, hours)
        else:
            return "{}'s duration is {} thus should have only bamboo leaves".format(self.name, hours)

panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female",3)
panda3 = Panda("Ming Ming", "Female",8)

print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))

print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))

print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep(13))
```

```python
#TASK_03
class GreenPhone:
    def __init__(self, model_name, android_version, num_cameras):
        self.model_name = model_name
        self.android_version = android_version
        self.num_cameras = num_cameras

    def updatePhone(self):
        if self.model_name.startswith('A') and self.android_version < 14:
            self.android_version += 1
            print(f"Your phone Greenphone {self.model_name} is upgraded to Android Version: {self.android_version}.")
        elif self.model_name.startswith('M') and self.android_version < 15:
            self.android_version += 1
            print(f"Your phone Greenphone {self.model_name} is upgraded to Android Version: {self.android_version}.")
        elif self.model_name.startswith('U') and self.android_version < 16:
            self.android_version += 1
            print(f"Your phone Greenphone {self.model_name} is upgraded to Android Version: {self.android_version}.")
        else:
            print(f"Your phone Greenphone {self.model_name} is already up to date.")

    def showSpecification(self):
        print("Phone Company: GreenPhone")
        print(f"Model Name: {self.model_name}")
        print(f"Android Version: {self.android_version}")
        print(f"Number of Cameras: {self.num_cameras}")

print('1=======================')
p1 = GreenPhone('A1', 12, 3)
p2 = GreenPhone('M11', 12, 4)
p3 = GreenPhone('U20', 12, 5)
p1.showSpecification()
print('2=======================')
p2.showSpecification()
print('3=======================')
p1.updatePhone()
print('4=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('5=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('6=======================')
p2.updatePhone()
p3.updatePhone()
print('7=======================')
p1.showSpecification()
p3.showSpecification()
```

```python
#TASK_04
class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}

    def calculateGPA(self, courses, semester):
        total_credits_attempted = len(courses) * 3
        total_grade_points = 0
        for course in courses:
            course_name, grade = course.split(': ')
            grade_point = float(grade)
            total_grade_points += grade_point * 3
        gpa = total_grade_points / total_credits_attempted
        if semester in self.grades:
            self.grades[semester].append(courses, gpa)
        else:
            self.grades[semester] = [(courses, gpa)]

    def printDetails(self):
        print(f'Name: {self.name}')
        print(f'ID: {self.id}')
        print('---------------------------------')
        for semester, courses_grades in self.grades.items():
            print(f'Courses taken in {semester}:')
            for courses, gpa in courses_grades:
                for course in courses:
                    print(course.split(': ')[0])
                print(f'CGPA: {gpa:.1f}')
                print('---------------------------------')

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```
